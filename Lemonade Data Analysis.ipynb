# Getting the data from CSV file

```python
import pandas as pd

df = pd.read_csv('Lemonade.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df
```

# Taking out the column values

```python
sales = df['Sales'].values
temp = df['Temperature'].values
rain = df['Rainfall'].values
flyers = df['Flyers'].values
price = df['Price'].values
```

# Doing the statistics of the given data

```python
import numpy as np
from scipy import stats
```

## Statistics for the sales

```python
sales_mean = np.mean(sales)
sales_median = np.median(sales)
sales_mode = stats.mode(sales)
sales_std = np.std(sales)
sales_var = np.var(sales)
print('Mean:',sales_mean)
print('Median:',sales_median)
print('Mode:',sales_mode)
print('Variance:',sales_var)
print('Standard Deviation:',sales_std)
```

## Statistics for the rainfall

```python
rain_mean = np.mean(rain)
rain_median = np.median(rain)
rain_mode = stats.mode(rain)
rain_std = np.std(rain)
rain_var = np.var(rain)
print('Mean:',rain_mean)
print('Median:',rain_median)
print('Mode:',rain_mode)
print('Variance:',rain_var)
print('Standard Deviation:',rain_std)
```

```python
temp_mean = np.mean(temp)
temp_median = np.median(temp)
temp_mode = stats.mode(temp)
temp_std = np.std(temp)
temp_var = np.var(temp)
print('Mean:',temp_mean)
print('Median:',temp_median)
print('Mode:',temp_mode)
print('Variance:',temp_var)
print('Standard Deviation:',temp_std)
```

# Visualization

```python
import matplotlib.pyplot as plt
```

## Histograms

```python
plt.figure(figsize=(15,10))
plt.hist(temp)
plt.title('Temperature(degree F)')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.hist(sales)
plt.title('Sales')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.hist(rain)
plt.title('Rainfall')
plt.show()
```

## Plotting graphs

### Sales-Temperature-Date Graph

```python
df['Temperature'].plot(figsize=(15,10))
df['Sales'].plot(figsize=(15,10))
plt.legend()
plt.title('Sales-Temperature-Date')
plt.show()
```

### Graph of Sales-Temperature

```python
plt.figure(figsize=(15,10))
plt.scatter(temp,sales,color='orange',marker='o')
plt.title('Temperaure-Sales')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.show()
```

### Rainfall-Sales Graph 

```python
plt.figure(figsize=(15,10))
plt.scatter(rain,sales,color='b',marker='o')
plt.title('Rainfall-Sales')
plt.xlabel('Rainfall')
plt.ylabel('Sales')
plt.show()
```

### Getting a linear relationship between rainfall and sales

```python
log_rain = np.log(rain)
plt.figure(figsize=(15,10))
plt.scatter(log_rain,sales,color='b',marker='o')
plt.title('Log_Rainfall-Sales')
plt.xlabel('Log_Rainfall')
plt.ylabel('Sales')
plt.show()
```

### Creating a graph between rain and date

```python
df['Rainfall'].plot(figsize=(15,10))
plt.title('Rainfall-Date')
plt.show()
```

### Creating subplots for better visualization and getting the relationship between the factors and output

```python
plt.subplot(4,2,1)
df['Rainfall'].plot(figsize=(15,20))
plt.title('Rainfall-Date')
plt.show()

plt.subplot(4,2,2)
df['Temperature'].plot(figsize=(15,20))
plt.title('Temperature-Date')
plt.show()

plt.subplot(4,2,3)
df['Sales'].plot(figsize=(15,20))
plt.title('Sales-Date')
plt.show()

plt.subplot(4,2,4)
df['Flyers'].plot(figsize=(15,20))
plt.title('Flyers-Date')
plt.show()
```

### Visualizing the number of flyers distributed and it's relation with sales

```python
plt.figure(figsize=(15,10))
plt.hist(flyers)
plt.title('Number of Flyers distributed')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(flyers,sales)
plt.title('Flyers-Sales')
plt.ylabel('Sales')
plt.xlabel('Flyers')
plt.show()
```

```python
print(np.corrcoef(rain,temp)) # getting the correlation coefficient between rainfall and temperature
```

```python
print(np.corrcoef(rain,sales)) # getting the correlation coefficient between rainfall and sales
```

```python
print(np.corrcoef(temp,sales)) # getting the correlation coefficient between temperature and sales
```

```python
print(np.corrcoef(flyers,sales)) # getting the correlation coefficient between flyers and sales
```

#### From the graphs, we can see that with increasing rainfall the sale of lemonade decreases exponentiatly. And from the Sales-Temperature graph we can see that the sales were comparatively higher when the temperature was high, i.e. on the hot days and during the cooler days the sales was low. During those days when rainfall was high we can see that temperature was low and vice-versa. The correlation coefficient was tells us the same thing. The correlation coeffiencient is negative which tells us rain and temperature are kind of inversely related. Sales and Rainfall are also inversely related as shown by the graph and the correlation coefficient. So from these analysis we can see that sales is higher on days with higher temperature and low rainfall, specially during June-July. Finally we can conclude that sales is higher on days when the number of flyers distributed is high, temperature is high and rainfall is low.
